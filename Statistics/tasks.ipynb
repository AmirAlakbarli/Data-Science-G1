```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# get distribution of age in dataset

df['age'].value_counts().sort_index().plot(kind='bar', figsize=(20,10))
```

```python
age_counts = df['age'].value_counts()
```

```python
mu = age_counts.mean()
mu
```

```python
sigma = np.std(age_counts)
sigma
```

```python
z_score = (age_counts - mu) / sigma
z_score
```

```python
z_score.plot(kind='hist', bins=100, figsize=(15, 5), title='Z-Score Distribution')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

z_scaled_data = scaler.fit_transform(age_counts.values.reshape(-1, 1))
z_scaled_data
```
